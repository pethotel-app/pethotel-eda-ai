```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터
years = np.arange(2007, 2023)
counts = np.array([14, 30, 37, 55, 93, 128, 195, 271, 340, 471, 577, 726, 902, 1132, 1480, 2012])

# 그래프 설정
fig, ax = plt.subplots(figsize=(12,6))

# x축, y축 라벨 설정
ax.set_xlabel('년도', fontsize=12)
ax.set_ylabel('반려동물 호텔 수', fontsize=12)

# x축 눈금 설정
ax.set_xticks(years)
ax.set_xticklabels(years, rotation=45, ha='right')

# y축 눈금 설정
y_ticks = np.linspace(0, 2500, 6)
ax.set_yticks(y_ticks)

# 라인 그래프 설정
ax.plot(years, counts, linewidth=2, marker='o', markersize=6, color='#006699')

# 막대 그래프 설정
for i in range(len(years)):
    if i == 0:
        ax.bar(years[i], counts[i], color='#006699', alpha=0.6, width=0.4)
    else:
        diff = counts[i] - counts[i-1]
        if diff > 0:
            ax.bar(years[i], diff, bottom=counts[i-1], color='#006699', alpha=0.6, width=0.4)

# 그래프 타이틀 설정
ax.set_title('대한민국 반려동물 호텔 등록 현황', fontsize=20)

# 그리드 설정
ax.grid(alpha=0.3)

# 보조선 설정
for y in y_ticks:
    ax.axhline(y=y, linestyle='--', alpha=0.3, color='black', linewidth=0.5)

# 그래프 보여주기
plt.show()
plt.show()
```
